<code>debit_reversal_success</code> columns.

<ul>
<li>
Missing values in the <code>td</code> column were filled using the median value 
to maintain numerical stability and reduce the effect of outliers.
</li>

<li>
Missing values in reversal-related columns were filled with 0 because the absence 
of reversal information may indicate that no reversal transaction occurred.

<h4 style="
    color:#0B3D91;">
    6.6 Handling Outliers and Invalid Value
    </h4>

```python
# Handle Invalid Approval %

print(f"Approval % below 0: {(df['approved'] < 0).sum()}")

print(f"Approval % above 100: {(df['approved'] > 100).sum()}")

df = df[
    (df['approved'] >= 0) &
    (df['approved'] <= 100)
]

# Handle Invalid TD %

df = df[
    (df['td'] >= 0) &
    (df['td'] <= 100)
]

# Handle Invalid BD %

df = df[
    (df['bd'] >= 0) &
    (df['bd'] <= 100)
]

# Remove Negative Transaction Volume

df = df[df['total_volume'] >= 0]

print("\nDataset Shape After Cleaning:", df.shape)
```

<div style="
background-color:#EAF4FF;
padding:12px;
border-left:6px solid #0B3D91;
border-radius:5px;">


Outlier analysis helps identify abnormal transaction behavior and unusual banking performance metrics within the dataset.    
Logical validation techniques were applied to remove invalid percentage values and negative transaction counts from the dataset. Since approval, technical decline, and business decline percentages must remain between 0 and 100, values outside this range were treated as inconsistent records and removed.

<h4 style="
    color:#0B3D91;">
    6.7 Creating Derived column 
    </h4>

```python
# Create Performance Score

df['Performance Score'] = (
    df['approved']
    - df['td']
    - df['bd']
)

# Create Risk Level

df['Risk Level'] = np.where(
    (df['td'] > 5) | (df['bd'] > 10),
    'High Risk',
    'Low Risk'
)


df['Performance Category'] = pd.cut(
    df['Performance Score'],
   bins=[0,70,90,100],
labels=['Poor','Average','Excellent']
)

df['Monthly Rank'] = df.groupby(
    ['fiscal_year', 'month']
)['approved'].rank(
    ascending=False,
    method='dense'
)
```

background-color:#EAF4FF;
padding:12px;
border-left:6px solid #0B3D91;
border-radius:5px;">

In this section, created new derived columns from existing 
data to improve analysis and extract meaningful insights. In this project, new 
features are created using banking transaction metrics such as approval percentage, 
technical decline, and business decline.

These derived columns help simplify transaction analysis, identify bank performance 
levels, and improve the understanding of UPI transaction patterns.


As part of feature engineering, I created four new columns. Performance Score combines approval rate, technical decline, and business decline to measure overall efficiency. Risk Level classifies banks based on technical decline percentage. Approval Category groups banks into Poor, Average, Good, and Excellent categories based on approval rates. Finally, Monthly Rank ranks banks within each month according to approval percentage, helping identify top-performing banks.

In this project, statistical analysis is performed on UPI transaction metrics

Skewness measures the asymmetry of the approval percentage distribution. It helps identify whether the data is positively or negatively skewed.

8.8 Approval Percentage vs Transaction Volume
</h3>

```python
comparison = (
    df.groupby('upi_remitter_banks')
      .agg({
          'approved': 'mean',
          'total_volume': 'mean'
      })
)
print(comparison.head())


plt.figure(figsize=(10,6))

sns.scatterplot(
    data=comparison,
    x='total_volume',
    y='approved'
)

plt.title('Approval Percentage vs Transaction Volume')

plt.xlabel('Average Transaction Volume')
plt.ylabel('Average Approval Percentage')

plt.show()
```
